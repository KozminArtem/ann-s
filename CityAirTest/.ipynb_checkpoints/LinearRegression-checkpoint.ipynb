# Вычисление концентрации газовых компонент через сверточный и линейный слой: часть 2

В первой части мы успешно реализовали нейросеть, устранющую воздействие приборной функции на спектр. Теперь настало время реализовать олносвязный слой, отвечающий за непосредственное вычисление концентраций

## Импорты

Начнем с импортирования сторонних библиотек:

```python
import os
from os.path import join
import json

import torch
from torch.utils.data import Dataset, random_split, DataLoader 
import torch.nn as nn
import torch.optim as optim

import numpy as np

from ray import tune
from ray.tune.schedulers import ASHAScheduler
```

## Формирование наборов данных

Создадим класс набора данных и функцию, формирующую тренировочный и тестовый набор:

```python
class LinearDataset(Dataset):
    def __init__(self, data_samples, data_targets, device='cuda:0'):
        self.samples = data_samples
        self.targets = data_targets
        self.device=device
        
    def __len__(self):
        return len(self.targets)
    
    def __getitem__(self, idx):
        sample = self.samples[idx]
        target = self.targets[idx]
        
        return torch.tensor(sample, device=self.device), torch.tensor(target, device=self.device)


def load_data(data_samples, data_targets):
    
    data = LinearDataset(data_samples, data_targets)

    # the constants to define
    test_rate = .2

    # the constants to calculate
    test_num = round(len(data)*test_rate)
    train_num = len(data) - test_num

    train_data, test_data = random_split(data, [train_num, test_num], generator=torch.Generator().manual_seed(42))   
    return train_data, test_data
```

## Нейронная сеть

Создадим класс нейронной сети только с теми параметрами, которые собираемся подбирать. В нашем случае это размер ядра свертки:

```python
class LinearLayer(nn.Module):
    def __init__(self, input_size):
        super(LinearLayer, self).__init__()
        
        self.fc = nn.Linear(input_size, 5)
    def forward(self, x):
        return self.fc(x)
```

## Функция тренировки

Добавим функцию тренировки с поддержкой (нескольких) видеокарт, точек сохранения а также автоматическим подгоном параметров:

```python
def train_fc(config, data_samples, data_targets, max_epochs=10, report=True):
    trainset, _ = load_data(data_samples, data_targets)
    
    fc = LinearLayer(data_samples.shape[1])
    
    device = "cpu"
    if torch.cuda.is_available():
        device = "cuda:0"
        if torch.cuda.device_count() > 1:
            fc = nn.DataParallel(fc)
    fc.to(device)
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(fc.parameters(), lr=config["lr"], 
                           weight_decay=config["weight_decay"])
    
    scheduler = optim.lr_scheduler.StepLR(optimizer, config["step_size"],
                                         config["gamma"])

    
    
    test_abs = int(len(trainset) * 0.8)
    train_subset, val_subset = random_split(
        trainset, [test_abs, len(trainset) - test_abs])
    
    trainloader = DataLoader(
        train_subset, 
        batch_size=32*torch.cuda.device_count(),
        shuffle=True)
    valloader= DataLoader(
        val_subset, 
        batch_size=32*torch.cuda.device_count(),
        shuffle=True)
    
    for epoch in range(max_epochs):
        running_loss = 0.0
        epoch_steps = 0
        
        
        for i, data in enumerate(trainloader):
            # get the inputs; data is a list of [sample, target]
            samples, targets = data
            
            # zero the parameter gradients
            optimizer.zero_grad()
            
            # forward + backward + optimize
            outputs = fc(samples)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            
            # print statistics
            running_loss += loss.item()
            epoch_steps += 1
   
    
        # Validation loss
        val_loss = 0.0
        val_steps = 0
        softmax = torch.nn.Softmax(dim=1)
        
        max_err = 0
        for i, data in enumerate(valloader):
            with torch.no_grad():
                samples, targets = data
            
                outputs = fc(samples)
                val_err = torch.max(torch.abs(targets - softmax(outputs)))
            
                loss = criterion(outputs, targets)
                val_loss += loss.cpu().numpy()
                val_steps += 1
                
                max_err = val_err if val_err > max_err else max_err
        
        scheduler.step()    
        
        if report:
            tune.report(loss=(val_loss / val_steps), max_err=float(max_err))
    if not report:
        print(f"Mean test error: {test_err(fc, data_samples, data_targets)}")
    return fc.state_dict()
```

## Ошбика на тестовых данных

Обычно точность машинного обучения измеряется на данных, которые не были использованы при тренировке моделей. Напишем для этого свою функцию:

```python
def test_err(fc, samples, targets, device='cpu'):
    _, testset = load_data(samples, targets)
    
    testloader = DataLoader(
        testset, batch_size=4, shuffle=False)
    err_acc = 0
    with torch.no_grad():
        softmax = torch.nn.Softmax(dim=1)
        for i_step, data in enumerate(testloader):
            samples, targets = data
            outputs = fc(samples)
            err_acc += torch.mean(torch.abs(softmax(outputs) - targets))
    return float(err_acc/i_step)
```

## Основное тело

Основная функция должна выглядить следующим образом:

```python
def main(num_samples=10, max_num_epochs=10, gpus_per_trial=2):

    data_dir = os.path.abspath("./data")
    data_ref = np.load(join(data_dir, 'raw+conv+mix.npz'))
    samples = data_ref['raw_conv'][0].astype(np.float32)
    targets = data_ref['gases'].astype(np.float32)
    

    config = {
        "lr": tune.loguniform(1e-7, 1e2),
        "weight_decay": tune.loguniform(1e-7, 1e-1),
        "step_size": tune.randint(1, 10),
        "gamma": tune.uniform(0, 1),
    }
    scheduler = ASHAScheduler(
        metric='loss',
        mode='min',
        max_t = max_num_epochs,
        grace_period =1, 
        reduction_factor=2)

    result = tune.run(
        tune.with_parameters(train_fc, data_samples=samples, data_targets=targets, max_epochs=max_num_epochs),
        resources_per_trial={'cpu':os.cpu_count(), 'gpu':gpus_per_trial},
        config=config,
        num_samples=num_samples,
        scheduler=scheduler)
    
    best_trial = result.get_best_trial("loss", "min", "last")
    print("Best trial config: {}".format(best_trial.config))
    print("Best trial final validation loss: {}".format(
        best_trial.last_result["loss"]))
    print(print("Best trial final max validation error: {}".format(
        best_trial.last_result["max_err"])))
    
    with open('res/best_config_fc.json', 'w') as f:
        json.dump(best_trial.config, f)
        
    print("Now training and save the best neural_network...")
    best_model_state = train_fc(best_trial.config, samples, targets, max_num_epochs,
                                  report=False)
    torch.save(best_model_state, 'res/best_state_fc.pt')
    

    
#if __name__ == "__main__":
#    main(num_samples=40, max_num_epochs=20, gpus_per_trial=torch.cuda.device_count())
```
